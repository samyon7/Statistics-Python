### Menghitung waktu yang dibutuhkan cahaya matahari untuk mencapai bumi setiap hari

Ada beberapa point acuan pada formula pada task kali ini.<br>

- Time = (Distance/Speed)
- r = (a(1-e**2)/1-e cos theta)

Dalam persamaan sebelumnya, nilai suatu adalah 149.600.000 km, yang merupakan jarak sumbu semi-major, e adalah 0,0167, yang merupakan eksentrisitas orbit bumi, dan Î¸ adalah sudut dari perihelion.

Dependent variable yang dibutuhkan adalah sebagai berikut<br>

theta = (2*phi*n) / 365.256

```python
import math
import numpy as np

# Two constant using capitalization
def earth_sun_distance():
    A = 1496e5
    E = 0.0167
    l = []

    # Calculate theta as it is different of each day of year. Let's calculate the distance of the earth from the sun
    for i in range(365):
        theta = (2*math.pi*i) / 365.25
        r = A*(1-E**2) / (1-(E*math.cos(theta)))
        l.append(r)
    return l
```

```python
# Assume the constant speed is 299.792 kms^-1
l = earth_sun_distance()
S = 299792
t = np.divide(l,S) # All value of L will divided by S
```

```python
sunny = np.asarray(list(zip(l,t)))
print(sunny)
```

### Matrix Search

Tugas kali ini kita akan mencari nilai input yang diberikan dalam matrix yang diurutkan dalam urutan naik, baik barisnya maupun kolomnya. Ini akan membantu kita memahami aturan umum traversal dalam matrix, terutama jika tidak menggunakan array NumPy. Tujuan akhirnya adalah mengembalikan nilai True atau False, tergantung pada apa nilai yang ada di dalam matriks tersebut

- Define first the matrix

```python
# Define first the matrix
matrix = [[7, 10, 15, 18],\
          [25, 29, 35, 47],\
           [56, 78, 85, 104]]
```

- Write a function, that will take this matrix as an input along with the value we have to search. We will first be covering the edge cases, which, in this instance, means where a matrix is empty or the target value is non-zero

```python
def matrixsearch(matrix, value):
    if value is None or not matrix:
        return False
    row = len(matrix)
    col = len(matrix[0])
    start = 0
    end = row*col -1
    while start <= end:
        mid = int((start+end)/2)
        pointer = matrix[int(mid/col)][int(mid%col)]
        print("Pointer :",pointer)
        print(int(mid/col),int(mid%col))
        if pointer == value:
            return True
        elif pointer < value:
            start = mid + 1
        else :
            end = mid-1
    return False


sol = matrixsearch(matrix,78)
print(sol)
```

### Use of Matrices in Performing Linear Equations

John is out of town for three days and in a mood to spend until he exhausts his resources. He has three denominations of currency with him. On the first day, John spends $435 on the latest electronic tablet that he likes, on which he spends 37 of type a denomination, 20 of type b, and 12 of type c. On the second day, he goes skydiving and spends 15, 32, and 4 of denominations a, b, and c, respectively, a total of $178. On the third day, with whatever amount he is left with, he decides to go to the theatre, which costs $70, and he shells out 5, 40, and 2 of the a, b, and c denominations, respectively. Can you tell what the values of the respective denominations are?

- Pertama, define variable 3  hari dalam matrix menggunakan array numpy

```python
# Ax = B

import numpy as np
z = np.array([[37,20,12],[15,32,4],[5,40,2]])
r = np.array([[435],[178],[70]])

# First calculate the inverse of matrix A
print(np.linalg.inv(z))

# Use the dot product an assign it into variable X
X = np.linalg.inv(z).dot(r)
print(X) # Solved

# This same thing can be done even more easily with the help of another NumPy function called solve(). Let's name the output variable here as y.
y = np.linalg.solve(z,r)
print(y) # Solved
```

Thus, we can see that the output after using both methods 1 and 2 is the same, which is $10, 25 cents, and $5, which are the respective denominations that we were trying to establish.

### Menemukan Probabilitas dari State Transitions

Anda diberikan sebuah array berisi 4 keadaan, yakni [A,B,C,D], di mana array ini dihasilkan secara acak. Selanjutnya, kita harus mencari probabilitas transisi 4 keadaan ini, di mana setelah ini, kita akan membentuk matrix transisi darinya. Mari implementasikan Markov Chains

- Gunakan library <i>random</i>. Setelah itu, tentukan berapa banyak elemen yang ingin kita jadikan sebagai konstanta, LEN_STR, yang mana dalam hal ini, akan kita atur menjadi 50

```python
import random

tokens = []
LEN_STR = 50
for i in range(LEN_STR):
    tokens.append(random.choice("ABCD"))

print(tokens)
```

- Temukan relative value, dan konversi mereka menjadi integer, di mana integer ini sendiri sangat mudah untuk kalkulasi secara cepat

```python
relative_value = [(ord(t) - ord('A')) for t in tokens]
print(relative_value)
```

- Sekarang, temukan perbedaan  antara ilai ASCII dan masukkan ke dalam sebuah list.

```python
LEN_TOKENS = len("ABCD")
m = [[0]*LEN_TOKENS for j in range(LEN_TOKENS)]
print(m)

for (i,j) in zip(relative_value, relative_value[1:]):
    m[i][j] += 1

print(list(zip(relative_value, relative_value[1:])))
print(m)
```

Di sini, kita sudah menginisiasi matrix nol berdasarkan ukuran LEN_TOKENS, yang kita define secara konstan untuk membangun matrix nol.

- Selanjutnya, kita akan membuat probabilitas, yang mana ini merupakan frekuensi relatif terhadap baris tertentu. Sebagai contoh, transisi dari A ke A adalah 8, dan total transisi dari A ke keadaan apapun adalah 20. Jadi, peluangnya adalah (8/20) = 0.4

```python
for state in m:
    total = sum(state)
    if total>0:
        state[:] = [float(f)/sum(state) for f in state]
```

- Cetak hasil statenya

```python
for state in m:
    print(state)
```

Dengan demikian, kita dapat membangun matriks transisi untuk menggambarkan transisi keadaan, yang menunjukkan kepada kita probabilitas transisi dari satu keadaan ke keadaan berikutnya. Jadi, kemungkinan A menemukan A sebagai huruf berikutnya adalah 0,4 , A menuju B adalah 0,15 , dan seterusnya

### Building a Text Predictor Using Markov Chain

Tujuannya adalah membangun prediktor teks kita sendiri. Pertama kita harus membuat list yang menjelaskan transisi status dengan membuat korelasi antar kata yang diberikan dan kata yang mengikutinya. Setelah itu, kita lakukan sorting pada list yang telah dibuyat, dan buat hash table dengan mengelompokkan kata yang mengikuti kata-kata tertentu pada posisi yang berbeda. Setelah itu gunakan random word generator untuk membuat dan mengarahkan value nya pada kata pertama.

Ada beberapa cara untuk menyelesaikan masalah ini, dan cara yang paling mudahlah yang akan kita ambil. Sebenarnya, memprediksi teks lebih sulit dari yang biasanya.

- Pertama, list kata kata di simpan dalam text file. Setelah itu, kita import dalam source code kita

```python
churchill = open('churchill.txt', encoding='utf-8').read()
keywords = churchill.split()
print(keywords)
```

- Iterasi list nya, dan tambahkan elements-element ini ke list yang baru, yang akan menyimpan keyword dan kata yang akan mengikutinya

```python
keylist = []
for i in range(len(keywords)-1):
    keylist.append((keywords[i], keywords[i+1]))
print(keylist)
```

- Selanjutnya, kita inisiasi sebuah dictionary word_dict. Setelah itu, kita iterasi lagi keylist yang sudah ada, dan menambahkan kata-katanya.

```python
word_dict = {}

for beginning, following in keylist:
    if beginning in word_dict.keys():
        word_dict[beginning].append(following)
    else:
        word_dict[beginning] = [following]

print(word_dict)
```

- Next, kita buat predictor kita sendiri. Pertama, define NumPy string, yang akan mengambil string secara acak sebagai pilihan dari kumpulan keywords sebelumnya

```python
first_word = np.random.choice(keywords)

while first_word.islower():
    first_word = np.random.choice(keywords)
```

```python
word_chain = [first_word]
```

```python
WORDCOUNT = 0
for i in range(WORDCOUNT):
    word_chain.append(np.random.choice(word_dict[word_chain[-1]]))
```

- Terakhir, define variable string <i>sentence</i>

```python
sentence = ' '.join(word_chain)
print(sentence)
```
